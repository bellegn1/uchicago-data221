```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats
```

```python
r = 0.5
a = scipy.stats.multivariate_normal.rvs(mean=[0,0], cov=np.array([[1, r], [r, 1]]), size=1000)
plt.plot(a[:,0], a[:,1], 'o')
a.shape
```

```python
means_1 = scipy.stats.multivariate_normal.rvs(mean=[0,1], cov=np.array([[1, 0], [0, 1]]), size=10)
means_2 = scipy.stats.multivariate_normal.rvs(mean=[1,0], cov=np.array([[1, 0], [0, 1]]), size=10)
```

```python
clusters_1 = (np.random.random(5000)*10).astype(int)
clusters_2 = (np.random.random(5000)*10).astype(int)
```

```python
cluster_means1 = means_1[clusters_1]
cluster_means2 = means_2[clusters_2]
cluster_means1.shape
```

```python
samples1x= scipy.stats.multivariate_normal.rvs( mean = cluster_means1[:,0] )
samples1y = scipy.stats.multivariate_normal.rvs( mean = cluster_means1[:,1] )

samples2x= scipy.stats.multivariate_normal.rvs( mean = cluster_means2[:,0] )
samples2y = scipy.stats.multivariate_normal.rvs( mean = cluster_means2[:,1] )
```

```python
plt.scatter(samples1x, samples1y, color="red")
plt.scatter(samples2x, samples2y, color="blue")
```
