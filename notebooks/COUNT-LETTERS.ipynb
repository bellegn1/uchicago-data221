```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats
from string import ascii_uppercase
```

```python
madhouse = open("madhouse.csv").read()
```

```python
clean = madhouse.replace("\n", " ").upper()
```

```python
#linux = open("Linux-FAQ.utf8").read()
#clean = linux.replace("\n", " ").replace("\l", " ")
```

```python
letters = {}
```

```python
for i in range(len(clean)):
    if clean[i] in letters.keys():
        letters[clean[i]] +=1
    else:
        letters[clean[i]] = 1
```

```python
# Turn letters dictionary into numpy vector
a = np.zeros(len(ascii_uppercase))
for i in range(len(ascii_uppercase)):
    if ascii_uppercase[i] in letters.keys():
        a[i] = letters[ascii_uppercase[i]]
    else:
        a[i] = 0
```

```python
# Build dictionary for counting bigrams:
h ={}
for i in ascii_uppercase+" ":
    for j in ascii_uppercase+" ":
        h[i+j] = 0 
len(clean)
```

```python
h.keys()
```

```python
# loop through the corpus, populate dictionary of bigrams
for i in range(len(clean)):
    k = clean[i:i+2]
    if k in h.keys():
        h[k] +=1
```

```python
h["AT"]
```

```python
alphabet = ascii_uppercase + " "
d = np.zeros((27,27))
for i in range(len(alphabet)):
    for j in range(len(alphabet)):
        bigram = alphabet[i]+alphabet[j]
        d[i][j] = h[bigram]
```

```python
#plt.bar(np.arange(26)+0.25, a, label=ascii_uppercase)
plt.bar(np.arange(27)-0.25, d.sum(axis=1)/d.sum(), color="red" )

plt.xticks(np.arange(27), labels=alphabet)
```

```python
# Plot heatmap of frequencies
plt.imshow(np.log(d), cmap="Blues")
plt.xticks(range(len(alphabet)), labels=list(alphabet))
plt.yticks(range(len(alphabet)), labels=list(alphabet))
```

```python
# Googling "matplotlib" "hinton diagram"
# https://matplotlib.org/3.1.1/gallery/specialty_plots/hinton_demo.html

def hinton(matrix, max_weight=None, ax=None):
    """Draw Hinton diagram for visualizing a weight matrix."""
    ax = ax if ax is not None else plt.gca()

    if not max_weight:
        max_weight = 2 ** np.ceil(np.log(np.abs(matrix).max()) / np.log(2))

    ax.patch.set_facecolor('gray')
    ax.set_aspect('equal', 'box')
#    ax.xaxis.set_major_locator(plt.NullLocator())
#    ax.yaxis.set_major_locator(plt.NullLocator())

    for (x, y), w in np.ndenumerate(matrix):
        color = 'white' if w > 0 else 'black'
        size = np.sqrt(np.abs(w) / max_weight)
        rect = plt.Rectangle([x - size / 2, y - size / 2], size, size,
                             facecolor=color, edgecolor=color)
        ax.add_patch(rect)

    ax.autoscale_view()
    ax.invert_yaxis()


if __name__ == '__main__':
    # Fixing random state for reproducibility
    np.random.seed(19680801)

    hinton(np.random.rand(20, 20) - 0.5)
    plt.title("Dummy hinton matrix")
    plt.show()
```
