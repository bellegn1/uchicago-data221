```python
import matplotlib.pyplot as plt
import numpy as np
import PIL
from sklearn.utils.extmath import randomized_svd
```

```python
filename = "3bearsplain.png"
def image_from_file(filename):
    i = PIL.Image.open(filename)
    Z = np.array(i)
    print(Z.shape)
    imagebw = np.sum(Z, axis=2) 
    imagebw = np.floor(imagebw / 3.0)
    print(imagebw.shape)
    return(imagebw)
```

```python
image1 = image_from_file("3bearsplain.png")
image2 = image_from_file("800px-UChicago_chimera_with_scarf.jpg")
```

```python
plt.imshow(image1, cmap="gist_gray");

image1.shape
```

```python
plt.imshow(image2, cmap="gist_gray")
```

```python
U, SIGMA, VT =randomized_svd(image1, n_components =10 )
approx = np.dot(np.dot(U, SIGMA*np.eye(U.shape[1])), VT)
```

```python
plt.imshow(approx, cmap="gist_gray")
```
