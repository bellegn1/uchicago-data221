```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# read penguins dataframe
df_train = pd.read_csv("penguins_train.csv")
df_test = pd.read_csv("penguins_test.csv")

sns.scatterplot(data=df_train,x='Flipper Length (mm)', y='Beak Length (mm)', hue='Species')

# clean data by removing nan entries
df_train = df_train[np.isfinite(df_train["Beak Depth (mm)"]) & 
                        np.isfinite(df_train["Body Mass (g)"]) &
                        np.isfinite(df_train["Flipper Length (mm)"]) & 
                        np.isfinite(df_train["Beak Length (mm)"])]

df_test = df_test[np.isfinite(df_test["Beak Depth (mm)"]) & 
                        np.isfinite(df_test["Body Mass (g)"]) &
                        np.isfinite(df_test["Flipper Length (mm)"]) & 
                        np.isfinite(df_test["Beak Length (mm)"])]

# one-hot encoding
df_train["SP1"] = df_train.Species == "Adelie"
df_train["SP2"] = df_train.Species == "Gentoo"
df_train["SP3"] = df_train.Species == "Chinstrap"
df_test["SP1"]= df_test.Species == "Adelie"
df_test["SP2"]= df_test.Species == "Gentoo"
df_test["SP3"]= df_test.Species == "Chinstrap"

# assign independent variables
X_train = df_train[['Beak Length (mm)', 'Beak Depth (mm)',    'Flipper Length (mm)',    'Body Mass (g)']]
X_test = df_test[['Beak Length (mm)', 'Beak Depth (mm)',    'Flipper Length (mm)',    'Body Mass (g)']]

# use sklearn package
from sklearn import linear_model

# train (fit) model for Gentoo
regr_genttoo = linear_model.LinearRegression()
regr_genttoo.fit(X_train.values, df_train.SP2.values)
coeff = regr_genttoo.coef_
intercept = regr_genttoo.intercept_

# find prediction (y-hat)
y = df_test.SP2
y_hat = np.dot(X_test.values, coeff) + intercept
y_hat

# find means
mean_bl0 = df_train['Beak Length (mm)'].mean()
mean_bd1 = df_train["Beak Depth (mm)"].mean()
mean_fl2 = df_train["Flipper Length (mm)"].mean()
mean_bm3 = df_train["Body Mass (g)"].mean() 

# plot line
x2div = np.arange(170, 240)
x0div = (0.5 - intercept - (coeff[1] * mean_bd1) - (coeff[3] * mean_bm3) - (coeff[2] * x2div)) / coeff[0]
plt.plot(x2div, x0div)
coeff
plt.ylim(20, 70        )
```
